# Sorting python pandas dataframe

```python
import pandas as pd
```

```python
mlb = pd.read_csv('data/mlbsalaries.csv')
```

```python
mlb.head()
```

```python
mlb.Year.value_counts()
```

```python
mlb.Team.value_counts()
```

```python
mlb.Position.value_counts()
```

```python
mlb.head()
```

```python
yr2010 = mlb[mlb.Year==2010]
```

```python
yr2010
```

```python
yr2010.values
```

```python
yr2010 = yr2010.set_index('Player')
```

```python
yr2010
```

```python
yr2010.head()
```

# sort by row labels

```python
yr2010.sort_index().head()
```

# sort by column labels

```python
yr2010.sort_index(axis=1).head()
```

# sort column values using order field, returns a Panda series

```python
yr2010.Salary.order(ascending=False).head()
```

# sort columns using sort_index() method, return the dataframe  

```python
sorted_yr2010 = yr2010.sort_index(ascending=False, by=['Salary'])
```

```python
sorted_yr2010.head()
```

```python
yr2010.sort_index(ascending = [False, True], by=['Salary', 'Team']).head()
```

# top 10 highest paid players

```python
top10 = yr2010.Salary.order(ascending=False).head(10)
```
