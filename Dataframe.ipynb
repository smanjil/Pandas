# The DataFrame object in pandas is a two-dimensional object.

```python
import datetime
```

```python
dt = datetime.datetime(2013,12,1)
end = datetime.datetime(2013,12,8)
step = datetime.timedelta(days=1)
dates = []
```

```python
while dt < end:
    dates.append(dt.strftime('%m-%d'))
    dt += step
```

```python
dates
```

```python
d = {'Date': dates, 'Tokyo': [15, 19, 15, 11, 9, 8, 13], 'Paris': [-2, 0, 2, 5, 7, -5, -3], 'Mumbai': [20, 18, 23, 19, 25, 27, 23]}
```

```python
d
```

# Create a dataframe object from a python dictionary of equal length lists
